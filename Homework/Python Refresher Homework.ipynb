# Lists

A **list** is used to store multiple items in a single variable.

### Key features of lists:
- They are written using square brackets: `[ ]`
- Items are separated by commas  
- Items have **indexes**, starting from 0. Thus the first item has index [0], the second item has index [1] etc.
- List indexing is therefore from 0 to n-1 where n is total number of elements in a list. For example a list of 100 items has index from 0 to 99 (n= 100, thus n-1 = 99)
- Lists are **changeable** (also known as mutable) ‚Äî you can update, add, or remove items

Examples:
```python
names = ["Jill", "Jack", "Jacob"]
scores = [10, 20, 30, 40]

````python
# Basic list example
colors = ["red", "green", "blue"]
print("Original list:", colors)
````

````python
# Accessing items
print("First color:", colors[0])
print("Last color:", colors[-1])
````

````python
# Changing an item
colors[1] = "yellow"
print("After change:", colors)
````

````python
# Adding items
colors.append("purple")
print("After append:", colors)
````

````python
# Length of list
print("Length:", len(colors))
````

````python
# Slicing examples
numbers = [10, 20, 30, 40, 50, 60, 70]

print("numbers:", numbers)
print("First 3 numbers:", numbers[:3])
print("Middle slice (index 2 to 4):", numbers[2:5])
print("Last 3 numbers:", numbers[-3:])

# List of lists
matrix = [[1,2,3], [4,5,6]]
print("Matrix:", matrix)
print("Element at row 2, column 2:", matrix[1][1])
````

# Dictionaries

A **dictionary** stores data in **key‚Äìvalue pairs**.

### Syntax:
```python
dictionary = {
    "key1": value1,
    "key2": value2
}


* Keys are unique
* You access values using the key
* Dictionaries are changeable
* Values can be any data type (string, number, list, another dictionary)


Example:

Common operations:
* Access value: person["name"]
* Add new key‚Äìvalue pair: person["college"] = "CTTE"
* Change value: person["age"] = 22
* Get all keys: person.keys()
* Get all values: person.values()

````python
# Basic dictionary example
person = {
    "name": "Jill",
    "age": 21,
    "city": "Chennai"
}

print("Dictionary:", person)
print("Name:", person["name"])
print("Age:", person["age"])
print("City:", person["city"])
````

````python
# Add a new key-value pair
person["college"] = "CTTE"
print("After adding college:", person)
````

````python
# Update a value
person["age"] = 22
print("After updating age:", person)
````

````python
# Keys and values
print("Keys:", list(person.keys()))
print("Values:", list(person.values()))
````

```python

Comparison operators:

Logical operators:

# Loops

Loops are used to execute a block of code several times as long as a certain criteria is met. if criteria is not met, loop is terminated and next statement outside the loop is executed

## While loop Refresher

A **while loop** executes a set of statements **as long as a condition is True**.

Basic pattern:

```python
i = 1
while i <= 5:
    # code block
    i += 1   # same as i = i+1
```

When the condition becomes `False`, the loop stops and the next statement after the loop is executed.


````python
i = 1
while i < 6:
    print(i)
    i += 1 # same as i = i+1
````

Note:
* With the 'break' statement we can stop a loop even if the while condition is true
* With the 'continue' statement we can stop the current iteration, and continue to the next iteration
* With the 'else' statement we can run a block of code when the loop condition is no longer true

````python
i = 1
while i < 6:
    print(i)
    i += 1
    if i == 3:
        continue # replace break with continue (or vice versa) and rerun code
else:
    print("i is no longer less than 6")
````


# Functions

A **function** is a reusable block of code that performs a specific task.

Functions make your code cleaner, modular, and easier to understand. 

### Function structure

```python
def function_name(parameters):
    # code here
    return result   # (optional)


## Writing a simple function and calling the function

````python
def print_lyrics():
    i = 1
    while i < 4:
        print("Baby shark, " + "do " * 6)
        i += 1
    print("Baby shark!")
````

````python
print_lyrics()
````

## Passing parameters to function Refresher

````python
def print_lyrics(name):
    i = 1
    while i < 4:
        print(f"{name} shark, " + "do " * 6)
        i += 1
    print(f"{name} shark!")
````

````python
print_lyrics("Baby")
print_lyrics("Mommy")
````

## Returning values from function Refresher

````python
# Function that returns a list of factors of x

def find_factors_of(x):
    factors = []
    for i in range(1, x+1):
        if x % i == 0:
            factors.append(i)
    return factors

print("Factors of 12:", find_factors_of(12))
print("Factors of 15:", find_factors_of(15))
````

## Nested functions

A function within a function forming a hierarchy. Inner function is only accessible inside outer function. 

````python
def outer_func(who):

    print(f"Hello")

    def inner_func():
        print(f"--{who}")

    inner_func()

outer_func("World!")
outer_func("Python!")
````

# Recursion

A **recursive function** is a function that calls itself.

It must have:
- a **base case** (when to stop)
- a **recursive step** (a smaller version of the original problem)

Example:  
Factorial  


````python
# Countdown example using recursion

def countdown(n):
    if n == 0:
        print("Blast off!")
    else:
        print("Counting down:", n)
        countdown(n - 1)

countdown(5)
````
